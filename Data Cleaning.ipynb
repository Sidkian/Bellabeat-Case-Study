# Data Processing

## Import Libaries and Data

```python
import numpy as np 
import pandas as pd
import os
```

```python
df_list = []
files = []

for dirname, _, filenames in os.walk('.\Raw Data'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        df = pd.read_csv(os.path.join(dirname, filename)) #Read each file into a dataframe
        df_list.append(df)                                  
        files.append(filename[:-4])
```

```python
i=0
for file in files:
    vars()[file] = df_list[i] #assigns filename to respective dataframe
    i+=1
```

```python
sleepDay_merged['SleepDay'] = pd.to_datetime(sleepDay_merged['SleepDay']).dt.strftime('%m/%d/%Y')
dailyActivity_merged['ActivityDate'] = pd.to_datetime(dailyActivity_merged['ActivityDate']).dt.strftime('%m/%d/%Y')
```
