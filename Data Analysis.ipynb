## Daily Data

### Heatmap of Daily Distance vs Minutes

```python
plt.figure(figsize=(20,10))
p1 = sns.heatmap(daily.corr(numeric_only=True)[['VeryActiveDistance','ModeratelyActiveDistance','LightActiveDistance']].loc['VeryActiveMinutes':'LightlyActiveMinutes'],
                 annot=True,
                annot_kws={'fontsize':15})
p1.tick_params(labelsize=15)
p1.set_title('Heatmap of Daily Distance vs Minutes',fontsize=30,pad=25)
```

### Scatter Plot of Daily Total Distance vs Calories

```python
plt.figure(figsize=(20,10))

r,p = sp.stats.pearsonr(x=daily['TotalDistance'],y=daily['Calories'])
p2 = sns.lmplot(x='TotalDistance',y='Calories',
           data=daily,aspect=2,height=8,
           scatter_kws={'color':'#58508d','s':45},
           line_kws={'color':'#ffa600'})
p2.axes[0,0].set_xlabel('Total Distance',fontsize=15)
p2.axes[0,0].set_ylabel('Calories',fontsize=15)
p2.axes[0,0].tick_params(labelsize=15)
p2.axes[0,0].set_title('Scatter Plot of Daily Total Distance vs Calories',fontsize=30,pad=25)
plt.text(25,5000,'r={:.2f}'.format(r),fontsize=15)
```

### Count of Daily Total Distance, Total Steps and Calories

```python
fig, ax = plt.subplots(1,3,figsize=(21,10))

p3 = sns.histplot(ax=ax[0],x='TotalDistance',data=daily,binwidth=5,color='#58508d')
p3.set_xlabel('Total Distance',fontsize=15)
p3.set_ylabel('Count',fontsize=15)
p3.tick_params(labelsize=15)
p3.set_title('Count of Daily Total Distance',fontsize=25,pad=25)

p4 = sns.histplot(ax=ax[1],x='TotalSteps',data=daily,binwidth=5000,color='#58508d')
p4.set_xlabel('Total Steps',fontsize=15)
p4.set_ylabel('Count',fontsize=15)
p4.tick_params(labelsize=15)
p4.set_title('Count of Daily Total Steps',fontsize=25,pad=25)

p5 = sns.histplot(ax=ax[2],x='Calories',data=daily,binwidth=1000,color='#58508d')
p5.set_xlabel('Calories',fontsize=15)
p5.set_ylabel('Count',fontsize=15)
p5.tick_params(labelsize=15)
p5.set_title('Count of Daily Calories',fontsize=25,pad=25)
```

### Monthly Usage

```python
monthly_usage = daily.groupby(by='Id').count()[['ActivityDate']]
monthly_usage.rename(columns={'ActivityDate':'DaysUsed'},inplace=True)
monthly_usage['Type'] = monthly_usage['DaysUsed'].apply(lambda x: 'High Use' if x > 20 
                                                        else ('Moderate Use' if x>10 and x<21 else 'Low Use'))
monthly_usage.head(10)
```

```python
fig, ax = plt.subplots(1,2,figsize=(20,10))

p6 = sns.histplot(ax=ax[0],
                  x='DaysUsed',
                  data=monthly_usage,
                  bins=3,
                  binrange=(0,31),
                  color='#58508d')
p6.set_xlabel('Days Used',fontsize=15)
p6.set_ylabel('Count',fontsize=15)
p6.tick_params(labelsize=15)
p6.set_title('Days Used in a Month',fontsize=25,pad=25)

ax[1].set_title('Monthly Usage Type Distribution',fontsize=25,pad=50)
_,_,autotext = ax[1].pie(data=monthly_usage.groupby('Type').count(),
                         x='DaysUsed',
                         labels=monthly_usage.groupby('Type').count().index,
                         labeldistance=1.2,
                         textprops={'fontsize':15},
                         autopct='%.2f%%',
                         colors=['#003f5c','#ffa600','#bc5090'])
for pct in autotext:
    pct.set_color('white')
    pct.set_fontsize(20)
```

### Daily Usage

```python
daily['TotalMinutes']=daily['VeryActiveMinutes']+daily['FairlyActiveMinutes']+daily['LightlyActiveMinutes']+daily['SedentaryMinutes']
daily_usage = daily[['Id','TotalMinutes']]

daily_usage.loc[daily_usage['TotalMinutes'] >=1200, 'Type'] = 'All Day'
daily_usage.loc[(daily_usage['TotalMinutes'] >= 600) & (daily_usage['TotalMinutes'] < 1200), 'Type'] = 'Half day'
daily_usage.loc[(daily_usage['TotalMinutes'] < 600), 'Type'] = 'Less than half day'
```

```python
fig, ax = plt.subplots(1,2,figsize=(20,10))
fig.suptitle('Daily Usage Type Distribution',fontsize=25)

p7 = sns.countplot(ax=ax[0],
                   x='Type',
                   data=daily_usage,
                   order=['Less than half day','Half day','All Day'],
                  palette = ['#ffa600','#bc5090','#003f5c'])
p7.set_xlabel('Type',fontsize=15)
p7.set_ylabel('Count',fontsize=15)
p7.tick_params(labelsize=15)


_,_,autotext = ax[1].pie(data=daily_usage.groupby('Type').count(),
                         x='Id',
                         labels=daily_usage.groupby('Type').count().index.sort_values(),
                         labeldistance=1.2,
                         textprops={'fontsize':15},
                         autopct='%.2f%%',
                         colors=['#003f5c','#ffa600','#bc5090'])
for pct in autotext:
    pct.set_color('white')
    pct.set_fontsize(20)
```

## Minute Data

```python
minute['METs'] = minute['METs']/10
minute
```

```python
condition = [minute['METs'] == 1.0,
            (minute['METs'] > 1.0) & (minute['METs'] <3),
            (minute['METs'] >= 3.0) & (minute['METs'] <=6),
             minute['METs'] > 6.0]
result = ['Sedentary','Light Intensity','Moderate Intensity','Vigorous Intensity']
minute['ActivityType'] = np.select(condition,result)
minute
```

### Total MET Minutes per person per week

```python
ActivityMinutes = minute.groupby(['Id','ActivityDate','ActivityType']).sum().reset_index()[['Id','ActivityDate','ActivityType','METs']]
ActivityMinutes.rename(columns={'METs':'METminutes'},inplace=True)
ActivityMinutes = ActivityMinutes[(ActivityMinutes['ActivityType'] == 'Moderate Intensity') | (ActivityMinutes['ActivityType'] == 'Vigorous Intensity')]

condition = [(ActivityMinutes['ActivityDate'] >= '04/11/2016') & (ActivityMinutes['ActivityDate'] <= '04/17/2016'),
             (ActivityMinutes['ActivityDate'] >= '04/18/2016') & (ActivityMinutes['ActivityDate'] <= '04/24/2016'),
             (ActivityMinutes['ActivityDate'] >= '04/25/2016') & (ActivityMinutes['ActivityDate'] <= '05/01/2016'),
             (ActivityMinutes['ActivityDate'] >= '05/02/2016') & (ActivityMinutes['ActivityDate'] <= '05/08/2016'),
             ActivityMinutes['ActivityDate'] >= '05/09/2016']
result = ['Week1','Week2','Week3','Week4','Week5']
ActivityMinutes['ActivityWeek'] = np.select(condition,result)

ActivityMinutes
```

```python
ActivityMinutes_Week = ActivityMinutes.groupby(['Id','ActivityWeek']).sum().reset_index()
ActivityMinutes_Week
```

```python
p14 = sns.barplot(data=ActivityMinutes_Week[ActivityMinutes_Week['ActivityWeek'] == 'Week1'],
                    x='Id',
                    y='METminutes',
                    color = '#58508d')
p14.set_xlabel('Id',fontsize=15)
p14.set_ylabel('MET Minutes',fontsize=15)
p14.tick_params(labelsize=10)
p14.tick_params(axis='x',rotation=90)
p14.set_title('MET Minutes per person (Week 1)',fontsize=25,pad=25)
rcParams['figure.figsize'] = 11,5
```

### MET Minute Type Distribution

```python
METminutesperweek = ActivityMinutes_Week.groupby(['Id']).mean().reset_index()

condition = [METminutesperweek['METminutes'] < 450 ,
            (METminutesperweek['METminutes'] >= 450) & (METminutesperweek['METminutes'] <900),
            (METminutesperweek['METminutes'] >= 900) & (METminutesperweek['METminutes'] <3000),
            (METminutesperweek['METminutes'] >= 3000) & (METminutesperweek['METminutes'] <=4000),
            METminutesperweek['METminutes'] > 4000]
result = ['Physically Inactive','Minimum MET Minutes','Good MET Minutes','Best MET Minutes','Extra MET Minutes']
METminutesperweek['Type'] = np.select(condition,result)

METminutesperweek
```

```python
METminutesperweek.groupby('Type').count()
```

```python
plt.figure(figsize=(10,10))
plt.title('MET Minute Type Distribution',fontsize=25,pad=25)

_,_,autotext = plt.pie(data=METminutesperweek.groupby('Type').count(),
                         x='Id',
                         labels=['Best MET Minutes','Extra MET Minutes','Good MET Minutes','Minimum MET Minutes'],
                         labeldistance=1.05,
                         textprops={'fontsize':15},
                         autopct='%.0f%%',
                        colors=['#003f5c','#bc5090','#ff6361','#ffa600'])

for pct in autotext:
    pct.set_color('white')
    pct.set_fontsize(15)
```

## Sleep Minutes

### Sleep Minute Distribution

```python
minuteSleep_value = sleep_minutes.groupby(['SleepDate','Id','value']).count().reset_index()
minuteSleep_value.groupby(['value']).mean().reset_index()[['value','SleepMinute']]
```

```python
plt.figure(figsize=(10,10))
plt.title('Average Sleep Minute Distribution',fontsize=25,pad=20)

_,_,autotext = plt.pie(data=minuteSleep_value.groupby(['value']).mean().reset_index()[['value','SleepMinute']],
                         x='SleepMinute',
                         labels=['Asleep (1)','Restless (2)','Awake (3)'],
                         labeldistance=1.05,
                         textprops={'fontsize':15},
                         autopct='%.2f%%',
                        colors=['#003f5c','#bc5090','#ffa600'])

for pct in autotext:
    pct.set_color('white')
    pct.set_fontsize(15)
```

### Sleep Type Distribution

```python
minuteSleep_byday = sleep_minutes.groupby(['SleepDate','Id']).count().reset_index()
minuteSleep_byday['Type'] = minuteSleep_byday['SleepMinute'].apply(lambda x: 'Good Sleep' if x>=420 else 'Bad Sleep')
minuteSleep_byday
```

```python
minuteSleep_type = minuteSleep_value.set_index(['SleepDate','Id']).join(minuteSleep_byday.set_index(['SleepDate','Id']), on=['SleepDate','Id'], how='left', rsuffix='r').reset_index()[['SleepDate','Id','value','Type','SleepMinute']]
minuteSleep_type
```

```python
minuteSleep_type[minuteSleep_type['Type'] == 'Good Sleep'].groupby(['value']).mean().reset_index()[['value','SleepMinute']]
```

```python
minuteSleep_type[minuteSleep_type['Type'] == 'Bad Sleep'].groupby(['value']).mean().reset_index()[['value','SleepMinute']]
```

```python
fig, ax = plt.subplots(1,2,figsize=(25,10))
fig.suptitle('Sleep Distribution by Sleep Type',fontsize=30)

ax[0].set_title('Good Sleep',fontsize=25)
_,_,autotext = ax[0].pie(data=minuteSleep_type[minuteSleep_type['Type'] == 'Good Sleep'].groupby(['value']).mean().reset_index()[['value','SleepMinute']],
                         x='SleepMinute',
                         labels=['Asleep (1)','Restless (2)','Awake (3)'],
                         labeldistance=1.05,
                         textprops={'fontsize':17},
                         autopct='%.2f%%',
                        colors=['#003f5c','#bc5090','#ffa600'])

for pct in autotext:
    pct.set_color('white')
    pct.set_fontsize(20)

ax[1].set_title('Bad Sleep',fontsize=25)
_,_,autotext = ax[1].pie(data=minuteSleep_type[minuteSleep_type['Type'] == 'Bad Sleep'].groupby(['value']).mean().reset_index()[['value','SleepMinute']],
                         x='SleepMinute',
                         labels=['Asleep (1)','Restless (2)','Awake (3)'],
                         labeldistance=1.05,
                         textprops={'fontsize':17},
                         autopct='%.2f%%',
                        colors=['#003f5c','#bc5090','#ffa600'])

for pct in autotext:
    pct.set_color('white')
    pct.set_fontsize(20)
```
